## Discretizing the State Space

### Assets

We use double-exponential transformation to get many points near the boundary. Let $[\underline{a},\overline{a}]=[0,10000]$. Then define $\overline{u} = \log(1 + \log(1 + \overline{a} - \underline{a})$ and construct the asset grid sequentially $a_i = \underline{a} + e^{e^{u_i}-1} - 1$.

```julia
function DiscretizeAssets(amin, amax, Na)
    ubar = log(1 + log(1 + amax - amin))
    ugrid = range(0, ubar, length = Na)
    agrid = zeros(length(ugrid))
    @tullio agrid[i] = amin + exp(exp(ugrid[i])-1) - 1
    return agrid
end
```

### Incomes

Assume $\log y_t$ follows an AR(1) process with persistence $\rho$ and normal innovations $\epsilon_t$

$$ \log y_t = \rho \log y_{t-1} + \epsilon_t $$

We will approximate this using the Rouwenhorst method (Kopecky and Suen) which is better than Tauchen-Hussey. 

```julia
function RouwenhorstP(N, p)
    """
    # Rouwenhorst method (Kopecky and Suen): Models sum of N-1 independent, binary states (0 or 1), 
    # each of which have probability p of staying a current value, 1-p of switching
    """
    # Base case 
    P = [p 1-p 
         1-p p]

    # Build Pn recursively 
    for n in 3:N
        P_old = copy.(P)
        P = zeros(n,n)

        P[1:end-1, 1:end-1] += p * P_old 
        P[1:end-1, 2:end] += (1-p) * P_old 
        P[2:end, 1:end-1] += (1-p) * P_old 
        P[2:end, 2:end] += p * P_old 
        P[2:end-1, :] /= 2
    end 
    return P
end
```

To match a given AR(1) process, we set $p = \frac{1+\rho}{2}$ and scale the sequence $s$ by $\alpha = \frac{2\sigma}{\sqrt{n-1}}$.

Furthermore, we seek the stationary distribution given by the transition matrix from Rouwenhorst:

```julia
function StationaryMarkov(P)
    """
    We know the eigenvalue associated with the stationary distribution is 1, so we impose 
    that the first entry in the stationary distribution is non-zero (1 in this case)
    and solve the resulting n-1 dimensional linear problem (I - P'[2:end])x = 0 which 
    is invertible, if the Ergodic theorem holds (i.e. P is aperiodic and irreducible <= Pij > 0)
    """
    Π = [1; (I - P'[2:end,2:end]) \ Vector(P'[2:end,1])]
    return Π = Π ./ sum(Π)
end
```

With these tools, we can now get the AR(1) approximation:

```julia
function DiscretizeIncome(ρ, σ, Ns)
    """
    # Match AR(1) process using Rouwenhorst: set p = (1+ρ)/2 and scale states s by α = 2σ/√(N-1)
    """
    p = (1+ρ)/2

    # Assume states are 0 to Ns-1, then scale using α 
    s = 0:Ns-1
    α = (2*σ)/sqrt(Ns-1)
    s = α .* s 

    # Markov transition matrix and stationary distribution 
    P = RouwenhorstP(Ns, p)
    Π = StationaryMarkov(P)

    # s is log income, get income y and scale so that mean is 1
    y = exp.(s)
    y /= dot(Π, y)

    return y, Π, P
end 
```

## EGM Method

### Step 1: Discounting and Expectations

Solving the household model in (1)-(3), we get 

$$ u'(c_t(s,a)) \geq \beta \mathbb{E}\Big[\frac{\partial}{\partial a'}V_{t+1}(s',a')\mid s\Big] =: \beta \mathbb{E}\big[V_{a, t+1}(s',a')\mid s\big] \tag{FOC} $$

Where the inequality is strict only when the borrowing constraint binds. To implement step 1, we guess $V_{a,t+1}$ and multiply with the transition matrix to get the expectation.

### Step 2: Solving for asset policy using the first order condition

If the inequality is strict, it means the borrowing constraint binds, i.e. $a' = \underline{a}$. When the inequality in (FOC) is an equality, we can invert it to get the EGM equation. Define $a_t(s,a')$ to be the inverse of $a_t'(s,a)$ such that $a_t(s,a_t'(s,a)) = a$. Then the EGM equation is

$$ c_t(s,a_t(s,a')) = (u')^{-1}\Big(\beta \mathbb{E}\big[V_{a,t+1}(s',a') \mid s\big]\Big) =: c_t^{endog}(s,a') \tag{EGM} $$

This is time-t consumption at the **endogenous gridpoint** $(s, a_t(s,a'))$, which is whatever asset level today that makes the household choose assets $a'$ tomorrow.

This consumption obeys the budget constraint $c_t^{endog}(s,a') + a' = (1+r_t)a + y_t(s) =: coh_t(s,a(s,a')) \tag{COH}$

and so to get asset policy $a'(a,s)$ we do the following: create a COH grid using the right-hand side of (COH) for all $(a',s) \in \mathcal{A} \times \mathcal{S}$. This gives $(coh(s,a_t(s,a'), a')$ which we can think of as a collection of sample points from the mappging that sends cash on hand into assets tomorrow. Linear interpolation of coh onto these $a'$ gives a mapping $(s,coh(s,a)) \mapsto a'$ i.e. our desired policy function $a'(s,coh(s,a))$ on the grid of $(s,a)$.

### Step 3: Enforcing borrowing constraint 

We just obtained $a_t(s,a)$ assuming the Euler equation holds with equality, ignoring the borrowing constraint. To fix this, we use that if Euler is an inequality, then the borrowing constraint binds and we can set $a_t'(s,a)=\underline{a}$: $a_t'(s,a) = max(a_t'(s,a),\underline{a})$.

We then back out consumption from the budget constraint as whatever is left over from cash on hand after we choose assets: $c_t(s,a)=coh_t(s,a)-a_t'(s,a)$. 

### Step 4: Use envelope condition to recover derivative of value function

Combining into one block of code, we get 

```julia
function EulerBack(Va, P, Agrid, y, r, β, σ)
    # 1. Discounted, expected value of Va tomorrow for all possible a', given today's state 
    Wa = (β .* P) * Va 

    # 2. Endogenous consumption from EGM equation
    c_endo = Wa.^(-σ)

    # 3. Cash-on-hand grid 
    coh = zeros(length(y), length(Agrid))
    @tullio coh[i,:] = y[i] .+ (1+r)*Agrid  

    a = copy.(coh)
    # interpolations[state] : c_endo[state,a'] + a' = coh ↦ a'
    interpolations = [linear_interpolation(c_endo[s,:] + Agrid, Agrid, extrapolation_bc = Line()) for s in 1:length(y)]
    # Evaluatng the interpolation at a coh[s,a] point implies an implicit a' which is collected in a[s,:]
    @tullio a[s, :] = interpolations[s](coh[s,:])

    # 4. Enforce borrowing constraint 
    a = max.(a, Agrid[1])
    c = coh - a

    # 5. Use envelop to get derivative of value function 
    Va = (1+r) * c.^(-1/σ)

    return Va, a, c
end
```

## Steady State Policies

To obtain steady state $V_a$ and policies $a',c'$, we need 

1. Initial guess of $V_a$
2. Iterate backward using EulerBack

To do this, we try $c$ equals 5% of cash on hand, use envelope to turn into guess of $V_a$

```julia
function PolicySS(P, Agrid, y, r, β, σ, tol=1e-9, maxiter=10000)
    # Initial guess for Va: consume 5% of cash-on-hand and get Va from envelope 
    coh = zeros(length(y), length(Agrid))
    @tullio coh[i,:] = y[i] .+ (1+r)*Agrid  
    c = 0.05*coh 
    Va = (1+r) * c.^(-1/σ)
    a_old = copy.(coh)

    # Iterate until maximum distance between two iterations fall below tol, or maxiter is reached 
    for it in 1:maxiter 
        Va, a, c = EulerBack(Va, P, Agrid, y, r, β, σ)
        if it > 1 && maximum(abs.(a - a_old)) < tol
            return Va, a, c 
        end 
        a_old = copy.(a)
    end
end
```

## Distribution 

Problem: Since we discretize assets, the asset policy is generally off-grid.

Solution: Lotteries. Suppose some asset policy $a'$ lies in between grid points $a_i,a_{i+1}$, then we want a lottery where the household goes to $a_i$ with probability $\pi$ and to $a_{i+1}$ with probability $1-\pi$ but we also want that when we aggregate up from the grid points, we get the true $a'$, i.e. 

$$ a' = \pi a_{i} + (1-\pi)a_{i+1} $$

which gives 

$$ \pi = \frac{a_{i+1} - a'}{a_{i+1}-a_i} $$

```julia
# Dealing with off-grid policies 
function GetLottery(a; Agrid)
    ss = searchsortedfirst(Agrid, a)                                # i such that A_i-1 a <= A_i
    if ss == 1     
        index = 1
        π = 1                                                       # If a' <= amin => assign all weight to amin 
    elseif ss >= length(Agrid)+1
        index = length(Agrid)-1
        π = 0                                                       # If a' >= amax => assign all weight to amax 
    else 
        index = ss - 1                                              # Index such that Agrid[index] <= a <= Agrid[index+1]
        π = (Agrid[index+1] - a)/(Agrid[index+1]-Agrid[index])      # Lottery probability
    end
    return index, π
end
```

```julia
# Apply to our example 
a_i, a_pi = GetLottery.(a; Agrid = a_grid)
```

Suppose we have a discrete distribution $D_t(s,a)$ with mass at each grid point. Then we send $\pi(s,a)$ of mass to gridpoint $i(s,a)$ and the rest to $i(s,a)+1$. This update beginning-of-period distribution $D_t(s,a)$ to end-of-period but before $s'$ draw distribution $D_t(s,a')$.

```julia
function ForwardPolicy(D, Lottery)
    """
    Given a discretized distribution D_t(s,a) with mass on each gridpoint, send π(s,a) mass to gridpoint
    i(s,a) and 1-π(s,a) to gridpoint i(s,a)+1. Return D_t^end(s,a') before the s' draw 
    I.e. hhs with (s,a) are sent to (s,a') by the policy, with Agrid[index] <= a' <= Agrid[index+1] for some index. 
    Use the Lottery to assign π(s,a) mass to (s,Agrid[index]) and 1-π(s,a) mass to (s,Agrid[index+1])
    """
    Dend = zeros(length(D[:,1]), length(D[1,:]))
    for s in 1:length(Lottery[:,1])
        for a in 1:length(Lottery[1,:])
            # Send π(s,a) of the mass to gridpoint i(s,a)
            Dend[s, Lottery[s,a][1]] += Lottery[s,a][2] * D[s,a]

            # Send 1-π(s,a) of the mass to gridpoint i(s,a)+1
            Dend[s, Lottery[s,a][1]+1] += (1-Lottery[s,a][2]) * D[s,a]
        end
    end
    return Dend
end
```

We need to iterate on the state as well. The previus function gets of from $D_t(s,a)$ to $D_t(s,a')$. Now we need to go from $D_t(s,a')$ to $D_{t+1}(s',a')$. This is a distribution, so we multiply by the transpose of the stationary distribution over states $\Pi'$.

```julia
function ForwardIteration(D, P, Lottery)
    """
    Input: Discretized distribution D_t(s,a), state transition matrix P and the Lottery.
    Output: D_t+1(s',a') i.e. the full iteration one period forward. ForwardPolicy tells us 
    (s,a') given (s,a). For each s, the hh can end up in one of length(y) states s' tomorrow 
    so the mass assigned to (s',a') should be the sum of all the points (si,a') * P(s' ∣ si)
    """ 
    Dend = ForwardPolicy(D, Lottery)
    return P' * Dend                            # Each masspoint Dend[s,a'] has probability P(s' ∣  s) of being in (s',a')
end
```

**Iterate to find steady-state distribution over (s,a)** Guess that $D(s,a)=\frac{1}{n_a}\pi(s)$, this ensures that the distribution over states is correct and imposes uniform distribution over assets in each state. 

```julia
function SSDistribution(P, apol, Agrid, tol=1e-10, maxiter=10000)
    Lottery = GetLottery.(apol; Agrid)
    Π = StationaryMarkov(P)
    # Initial guess: stationary over y, uniform over a 
    D = ones(length(P[:,1]),length(Agrid)) 
    @tullio D[i,:] = (Π[i] / length(Agrid)) * D[i,:]

    # Iterate until convergence 
    for it in 1:maxiter
        D_new = ForwardIteration(D, P, Lottery)
        if maximum(abs.(D_new - D)) < tol 
            return D_new 
        end 
        D = copy.(D_new)
    end
    return D
end
```

## Aggregating to full steady state

```julia
# Combine part 2 and 3 into one function
function SteadyState(P, Agrid, y, r, β, σ)
    Va, apol, cpol = PolicySS(P, Agrid, y, r, β, σ)
    D = SSDistribution(P, apol, Agrid)

    return (; D = D, Va = Va, apol = apol, cpol = cpol, A = dot(apol, D),
             C = dot(cpol, D), P = P, Agrid = Agrid, y = y, r = r, β = β,
             σ = σ)
end
```

### Application: Comparative Statics in r

```julia
# Comparative statics in r: Steady state assets 
function CompStatR(r,nr,rmin,rmax,Agrid,P,y,β,σ)
    rs = r .+ range(rmin, rmax, length=nr)
    Ar = zeros(nr)
    Threads.@threads for i in 1:nr 
        Ar[i] = SteadyState(P, Agrid, y, rs[i], β, σ).A
    end 
    return rs, Ar
end
rs,Ar = CompStatR(r,15,-0.02,0.015,a_grid,Pi,y,β,σ)
plot(rs, Ar, title="Comparative Statics in r", 
     ylabel="Aggregate assets over quarterly income", xlabel = "Real interest rate",
     legend=false)
vline!([β^(-1)-1])
```

### Application: Comparative statics in income risk

```julia
# We manipulate income risk by changing σ in the income process
# As income risk increases, assets increase for precautionary reasons  
function CompStatSigma(sigma_low,sigma_high,ns,Agrid,r,β,σ)
    σincs = range(sigma_low,sigma_high,length=ns)
    incomes = [DiscretizeIncome(0.975, i, 7) for i in σincs]
    Ar = zeros(ns)
    Threads.@threads for i in 1:ns
        Ar[i] = SteadyState(incomes[i][3], Agrid, incomes[i][1], r, β, σ).A
    end 
    return σincs, Ar
end
σincs, Ar = CompStatSigma(0.3,1.2,8,a_grid,r,β,σ)
plot(σincs, Ar, title="Comparative Statics in income risk", 
     ylabel="Assets over quarterly income", xlabel = "Cross-sectional standard dev. of income",
     legend=false)
```

### Application: Comparative statics in intertemporal substitution

```julia
# Comparative statics in intertemporal substitution 
# Assets are decreasing in EIS, or equivalently, increasing in risk aversion
# The more risk averse you are, the more you want to smooth consumption intertemporally
# The higher the elasticity of inter-temporal substitution, the more willing you are to 
# vary consumption over time and so you need fewer assets to dampen shocks and 
# assets held are assets not converted into consumption 
function CompStatEis(sigma_low,sigma_high,ns,Agrid,P,y,r,β)
    σstemp = range(sigma_low,sigma_high,length=ns)
    Ar = zeros(ns)
    Threads.@threads for i in 1:ns 
        Ar[i] = SteadyState(P, Agrid, y, r, β, σstemp[i]).A
    end
    return σstemp, Ar
end
σstemp, Ar = CompStatEis(0.4,2.0,10,a_grid,Pi,y,r,β)
plot(σstemp, Ar, title="Comparative Statics in EIS", 
     ylabel="Assets over quarterly income", xlabel = "Elasticity of Intertemporal Substitution",
     legend=false)
```

## Calibration

Inspired by (McKay, Nakamura, Steinsson) we want to calibrate the model such that 
total assets are 140% of annual GDP. Since income in this model is quarterly, we want 
A such that 4*1.4 = A = 5.6. We allow β to vary in order to do this 

```julia
function BetaCalib(P,Agrid,y,r,σ,beta_low,beta_high,Assetlevel)
    β_calib = find_zero(x ->  SteadyState(P, Agrid, y, r, x, σ).A - Assetlevel, [beta_low beta_high])
    ss = ss = SteadyState(P, Agrid, y, r, β_calib, σ)
    A, C = ss.A, ss.C
    # Check aggregate steady-state budget balance: C = 1+rA 
    if isapprox(C, 1+r*A) == 1
        println("Calibration was succesful")
    else 
        println("Calibration failed")
    end
    return β_calib
end
β_calib = BetaCalib(Pi,a_grid,y,r,σ,0.98,0.995,5.6)
```

## General Equilibrium

So far, we have only considered household problem in isolation (partial eq). We turn it into a GE setting by adding a production sector as well as the opportunity for households to save in government bonds, the supply of which satisfies a budget constraint. 

1. Households have stochastic endowments $e(s)$ of effective labor, following the same markov process $y(s)$ as before
2. Each unit produces on unit of consumption goods in a perfectly competitive production sector, hence real wage is 1
3. HHs save in government bonds, paying real interest rate $r$ and supplied in exogenous quantity $B$
4. Government finances this with tax $\tau = rB$ on laobr, hence HHs after-tax labor income is $y(s)=(1-\tau)e(s)$

So the only difference from before is that now the households' budget constraint involves $1-\tau$.

To solve it, we vary income risk, keeping everything else constant. We vary r to satisfy asset market clearing for 
each income risk. r enters the HH problem as a separate variable and in the government budget constraint. 
This means that the tax changes with r. As income risk increases, household want to save more and so
r falls to make sure that the market still clears. 

```julia
B = 5.6                     # Bonds supplied by government, paying real interest r 
τ = r*B                     # Labor tax to balance steady-state government budget 
e = y
function BetaCalibGE(Pi, Agrid, τ, e, r, σ, B)
    β_ge = find_zero(x ->  SteadyState(Pi, Agrid, (1-τ)*e, r, x, σ).A - B, [0.98 0.995])    # Calibrate β to be consistentwith asset market clearing A = B
    ss_ge = SteadyState(Pi, Agrid, (1-τ)*e, r, β_ge, σ)
    A_ge, C_ge = ss_ge.A, ss_ge.C

    if isapprox(C_ge, 1) == 1 & isapprox(A_ge, B) == 1
        println("Calibration succesful")
    else 
        println("Calibration unsuccesful")
    end
    return β_ge
end
β_ge = BetaCalibGE(Pi, a_grid, τ, e, r, σ, B)
```

### General equilibrium counterfactuals

```julia
# As income risk increases, so does the demand for assets to help smooth consumption.
# This brings down the return on assets. 
function GECounter(β_ge, Agrid, B, σ)
    σincs = range(0.3,1.2,length=8)
    r_ge = zeros(8)
    Threads.@threads for i in 1:length(σincs)
        println(i)
        ei, Πi, Pi = DiscretizeIncome(0.975, σincs[i], 7)
        r_ge[i] = find_zero(x ->  SteadyState(Pi, Agrid, (1-x*B)*ei, x, β_ge, σ).A - B, [-0.02 0.015])
    end
    return r_ge, σincs
end
rge, σincs = GECounter(β_ge, a_grid, B, σ);
```

## j-period-ahead Expectations

Suppose we have some function $x(s,a)$ of the state. If the state is $(s_t,a_t)$ today what do we expect $x(s_{t+j},a_{t+1})$ to be? This is given by the $j$th expectation vector 

$$ \mathcal{E} = \mathbb{E}[x(s_{t+j},a_{t+j}) \mid (s_t,a_t)] $$

by LIE, we can do this recursively if we define $\mathcal{E}_0(s,a)=x(s,a)$: 

$$ \mathcal{E}_j(s,a) = \mathbb{E}[\mathcal{E}_{j-1}(s',a') \mid (s,a)] $$

So as long as we can calculate one-period ahead expectations, we can calculate expectations over any horizon!

**Illustration** Let's take a simple example and start by assuming that the recursion is correct, which means that the following is true 

$$ \mathcal{E}_2(s,a) = \mathbb{E}[x(s'',a'') \mid (s,a)] = \mathbb{E}[ \mathcal{E}_1(s',a') \mid (s,a)] \tag{True?} $$

The first equality is the definition, the second equality is what we start by just accepting as being true. If (True?) is true, we can apply LIE recursively to get 

$$ \mathcal{E}_2(s,a) = \mathbb{E}[ \mathcal{E}_1(s',a') \mid (s,a)] = \mathbb{E}[ \mathbb{E}[ \mathcal{E}_0(s'',a'')\mid (s',a')] \mid (s,a)] = \mathbb{E}[ \mathbb{E}[ x(s'',a'')\mid (s',a')] \mid (s,a)] $$

where we use the definition of $\mathcal{E}_0$ in the end.

So we first calculate $ \mathbb{E}[ x(s'',a'')\mid (s',a')] $ in two steps. First, for each $a'' \in A$ we calculate the conditional expectation over $s'$

$$ E[x(s'',a'') \mid s'] = \sum_{s'' \in S} P(S = s'' \mid S = s')
x(s'',a'') = X^{end}(s',a'') = f(s',a'') $$

Secondly, using these conditional expectations, we take conditional expectations with respect to $a'$

$$ \mathbb{E}[ x(s'',a'')\mid (s',a')] = \sum_{a'' \in A} P(A = a'' \mid A = a') X^{end}(s',a'') $$

$$ = \pi(s',a')X^{end}(s',i(s',a')) + (1-\pi(s',a'))X^{end}(s',i(s',a')+1) $$

$$ = \pi(s',a')E[x(s'',i(s',a') \mid s'] + (1-\pi(s',a'))E[x(s'',i(s',a')+1) \mid s']$$

$$ =: X(s',a')  $$

Where we use the notation from the code $\pi(s',a')$ is the mass assigned to $a''$ and $1-\pi(s',a')$ is the mass assigned to the the grid point one above $a''$. This accounts for the fact that the policy is off grid. These two steps are implemented by passing the policy to the function $ExpectationIteration()$ in the code. To get $\mathcal{E}_0$ we just repeat the procedure once more.

Finally, why is $(True?)$ true? 

### Take expectations with respect to the lottery

We start with end-of-perid $X_t^{end}(s,a')$ and obtain its beginning of period expectation $X_t(s,a)=\pi(s,a) \cdot X_t^{end}(s,i(s,a)) + (1-\pi(s,a))\cdot X_t^{end}(s,i(s,a)+1)$

```julia
function ExpectationPolicy(Xend, ai, a_pi)
    """
    Goes from X_t^(end)(s,a') - before an s' draw - to X_t(s,a), i.e. the opposite way of ForwardPolicy. 
    Takes a lottery - i.e. all elements a_i that are being mapped to tomorrow from states today along with the weight given to them - 
    and creates the expected value today. The intuition is this: a policy x(s,a) maps (s,a) to something off grid x(s,a) = (s,a')
    that lands between two grid points a_i and a_(i+1), with weight a_pi(a,s) and (1-a_pi(a,s)) respectively. 
    Now, take any grid point tomorow a_i that is hit by x(s,a) for some (s,a). We know that a_i(s,a) <= x(s,a) < a_(i+1)(s,a) and that 
    each grid point is assigned weights a_pi(a,s) and (1-a_pi(a,s)) respectively, so we can calculate the expected value of 
    x(s,a) as the mixture of a_i(s,a) and a_(i+1)(s,a). I think this expectation is actually identical to x(s,a), but it will depend
    on how the weights are constructed.
    """
    X = zero(Xend)
    for s in 1:length(a_i[:,1])
        for a in 1:length(ai[1,:])
            X[s,a] = a_pi[s,a] * Xend[s, a_i[s,a]] + (1-a_pi[s,a])*Xend[s, a_i[s,a]+1]
        end
    end
    return X
end
```

### Combine with expectations of the state 

Recall that for the distribution, we updated $D_t^{end}(s,a')$ to $D_{t+1}{end}(s',a')$ by mupltiplying by the transpose of the Markov matrix. Now, we need to obtain $X_t^{end}(s,a')=\mathbb{E}_t[X_{t+1}(s',a')\mid s]$ by taking expectations with respect to $s'$, then combine with the code above to obtain $X_t(s,a)$ from $X_t^{end}(s,a')$. This gives

```julia
function ExpectationIteration(X, Pi, a_i, a_pi)
    """ 
    ExpectationPolicy takes X_t^(end)(s,a') and returns X_t(s,a). We now need to obtain X_t^(end)(s,a') from X_t+1(s',a'). 
    This is done by taking expectations, similarly to ExpectationPolicy which is also an expectation. 
    Note how to go forward we used P' to get distributions, so to reverse these operations, we take expectations using P. 
    Px gives a vector of conditional expectations. P'x gives a distribution. 
    Also note that the order of operations is reversed. Before, we did ForwardPolicy(s,a) ↦ (s,a') then 
    ForwardIteration(s,a') ↦ (s',a'). Now, we go from ExpectationIteration(s',a') ↦ (s,a'), then ExpectationPolicy(s,a') ↦ (s,a) 
    because we have to undo the operations in order. 
    """
    Xend = Pi * X
    return ExpectationPolicy(Xend, a_i, a_pi)
end
```

Write a function that takes in any function $x(s,a)$ of the state and calculate expectation vectors $\mathcal{E}_0,...\mathcal{E}_T-1$

```julia
function ExpectationVectors(X, Pi, a_i, a_pi, T)
    curlyE = [zero(X) for t in 1:T]
    curlyE[1] = X 

    for j in 2:T
        curlyE[j] = ExpectationIteration(curlyE[j-1], Pi, a_i, a_pi) 
    end
    return curlyE
end
```

With this function in hand, we can calculate e.g. auto-correlation of consumption. See https://conference.nber.org/confer/2022/MWs22/Lecture1.html for more detailes 

```julia
## Autocorrleation of consumption 
ctilde = c .- dot(λss,c)             # Demeaned consumption 
a_i = [GetLottery(a; Agrid = a_grid)[1] for a in a]
a_pi = [GetLottery(a; Agrid = a_grid)[2] for a in a]
E_ctilde = ExpectationVectors(ctilde, Pi, a_i, a_pi, 40)

Autocov_c = zeros(40)
for j in 1:40
    Autocov_c[j] = dot(λss, ctilde .* E_ctilde[j])     # ctilde .* E_ctilde[j] is the product of consumption in state (s,a) today with consumption in state (s,a) tomorrow. Multiply with distribution to get average autocovariance
end
# Autocorrelation is autocovariance divided by variance, and variance is autocovariance with lag 0
Autocorr_c = Autocov_c ./ Autocov_c[1]

plot(1:40, Autocorr_c, title = "Autocorrelation of consumption", ylabel = "Correlation", xlabel = "Horizon in quarters")
```
